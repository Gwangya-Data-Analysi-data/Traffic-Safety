```python
import requests
import geopandas as gpd
import io

# 설정값
url = "https://www.safemap.go.kr/openApiService/wms/getLayerData.do"
params = {
    "apikey": api_key,
    "layername": "A2SM_CRMNLHSPOT_TOT",
    "styles": "A2SM_CrmnlHspot_Tot_Tot"
}

# 요청
response = requests.get(url, params=params)

# 응답 확인
if response.status_code == 200:
    try:
        print("📄 응답 샘플:\n", response.text[:500])  # 응답 앞부분 출력

        # GeoDataFrame으로 읽기 (GML로 가정)
        gdf = gpd.read_file(io.StringIO(response.text))

        # 저장
        gdf.to_excel("범죄주의구간_전체_GML.xlsx", index=False)
        print("✅ 엑셀 저장 완료!")
    except Exception as e:
        print("⚠️ 데이터 파싱 중 오류:", e)
else:
    print("❌ API 요청 실패:", response.status_code)
```

```python
import requests

# ▶️ 서비스 파라미터
param = {
    "name": "범죄주의구간(전체)",
    "serverUrl": "https://www.safemap.go.kr/openApiService/wms/getLayerData.do",
    "apikey": "YOUR_API_KEY_HERE",
    "layername": "A2SM_CRMNLHSPOT_TOT",
    "styles": "A2SM_CrmnlHspot_Tot_Tot"
}

# ▶️ GetMap 요청 매개변수 (좌표계·범위·해상도는 필요에 맞게 수정)
bbox      = "126.8,37.45,127.2,37.75"      # (xmin,ymin,xmax,ymax) ‑ EPSG:4326
img_size  = (800, 600)                      # width, height
base_qs   = {
    "SERVICE": "WMS",
    "VERSION": "1.3.0",
    "REQUEST": "GetMap",
    "CRS": "EPSG:4326",
    "BBOX": bbox,
    "WIDTH": img_size[0],
    "HEIGHT": img_size[1],
    "LAYERS": param["layername"],
    "STYLES": param["styles"],
    "FORMAT": "image/png",
    "TRANSPARENT": "TRUE"
}

# ▶️ 호출
resp = requests.get(
    f'{param["serverUrl"]}?apikey={param["apikey"]}',
    params=base_qs,
    timeout=30
)
resp.raise_for_status()

# ▶️ 파일로 저장
with open("crime_hotspots.png", "wb") as f:
    f.write(resp.content)

print("✔️ crime_hotspots.png 저장 완료")
```

```python
import folium


APIKEY  = "QX7GV3XP-QX7G-QX7G-QX7G-QX7GV3XP1P"

wms_url = (
    "https://www.safemap.go.kr/openApiService/wms/getLayerData.do?"
    f"apikey={APIKEY}"
)

m = folium.Map(location=[37.5665, 126.9780], zoom_start=11, tiles="CartoDB positron")

folium.raster_layers.WmsTileLayer(
    url=wms_url,
    layers="A2SM_CRMNLHSPOT_TOT",
    styles="A2SM_CrmnlHspot_Tot_Tot",
    fmt="image/png",
    transparent=True,
    name="범죄주의구간(전체)",
    attr="SafeMap",
).add_to(m)

folium.LayerControl().add_to(m)
m  # 노트북이면 지도가 바로 표시됩니다

```
