```python
import requests

# â–¶ï¸ ì„œë¹„ìŠ¤ íŒŒë¼ë¯¸í„°
param = {
    "name": "ë²”ì£„ì£¼ì˜êµ¬ê°„(ì „ì²´)",
    "serverUrl": "https://www.safemap.go.kr/openApiService/wms/getLayerData.do",
    "apikey": "YOUR_API_KEY_HERE",
    "layername": "A2SM_CRMNLHSPOT_TOT",
    "styles": "A2SM_CrmnlHspot_Tot_Tot"
}

# â–¶ï¸ GetMap ìš”ì²­ ë§¤ê°œë³€ìˆ˜ (ì¢Œí‘œê³„Â·ë²”ìœ„Â·í•´ìƒë„ëŠ” í•„ìš”ì— ë§ê²Œ ìˆ˜ì •)
bbox      = "126.8,37.45,127.2,37.75"      # (xmin,ymin,xmax,ymax) â€‘ EPSG:4326
img_size  = (800, 600)                      # width, height
base_qs   = {
    "SERVICE": "WMS",
    "VERSION": "1.3.0",
    "REQUEST": "GetMap",
    "CRS": "EPSG:4326",
    "BBOX": bbox,
    "WIDTH": img_size[0],
    "HEIGHT": img_size[1],
    "LAYERS": param["layername"],
    "STYLES": param["styles"],
    "FORMAT": "image/png",
    "TRANSPARENT": "TRUE"
}

# â–¶ï¸ í˜¸ì¶œ
resp = requests.get(
    f'{param["serverUrl"]}?apikey={param["apikey"]}',
    params=base_qs,
    timeout=30
)
resp.raise_for_status()

# â–¶ï¸ íŒŒì¼ë¡œ ì €ì¥
with open("crime_hotspots.png", "wb") as f:
    f.write(resp.content)

print("âœ”ï¸ crime_hotspots.png ì €ì¥ ì™„ë£Œ")
```
